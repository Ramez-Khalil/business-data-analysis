# Business Data Analysis Notebook

This notebook demonstrates basic data analysis techniques applied to a business dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [12000, 15000, 17000, 16000, 18000]
}

df = pd.DataFrame(data)
df
```

```python
plt.plot(df["Month"], df["Sales"])
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.show()
```
